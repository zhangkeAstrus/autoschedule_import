```python
import requests
import pandas as pd
import urllib3

# Disable SSL warnings
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# VIN list
vin_list = [
    "1FTFW1RG1RFA12345",
    "1FTEW1EF5GFB60292",
    "1FTFW1E88PFB69817",
    "1FTFX1E53MFB64104",
    "1FTEX1EBXKKD42015",
    "1FTFW1E59PFA61504",
    "1FTFW5LD0SFA01220",
    "1FTEW2KP8RKE28703",
    "1FTEW2LP7RKE66812",
    "3TMCZ5ANXNM475346",
    "3TMCZ5AN1NM504328",
    "3TMCZ5AN1NM479785",
    "5TFRX5GNXKX160656",
    "5TEJU62N15Z074717",
    "3TMCZ5AN5MM403419",
    "3TMJU62N79M074112",
    "3TMKB5FN3SM034257",
    "3TMKB5FN1SM034662",
    "3TMLB5JN7SM127049",
    "3TMLB5JN0SM117091",
    "3TMLB5JN1SM123725"
]

# Store all results
all_records = []

for vin in vin_list:
    url = f"https://vpic.nhtsa.dot.gov/api/vehicles/DecodeVinValuesExtended/{vin}?format=json"
    try:
        response = requests.get(url, verify=False)
        data = response.json()
        result = data['Results'][0]
        all_records.append(result)
    except Exception as e:
        print(f"Error decoding VIN {vin}: {e}")

# Convert to DataFrame
df = pd.DataFrame(all_records)

# Display the full result
df.head()
```
